```python
from bs4 import BeautifulSoup

# Функция для рекурсивного обхода категорий и записи их в файл
def parse_categories(dl_tag, output_file, level=0):
    print(f"Обработка <DL> на уровне {level}")  # Отладочный вывод
    # Находим все элементы внутри текущего <DL>
    for element in dl_tag.find_all(recursive=False):
        print(f"Обнаружен элемент: {element.name}")  # Отладочный вывод
        if element.name == 'p':  # Если это <p>, ищем внутри него <DT> и <H3>
            print("Найден элемент <p>, проверяем его содержимое")
            dt_tag = element.find('dt')
            if dt_tag:  # Если внутри <p> есть <DT>
                h3_tag = dt_tag.find('h3')  # Ищем <H3> внутри <DT>
                a_tag = dt_tag.find('a')  # Ищем <A> внутри <DT>
                if h3_tag:  # Если нашли <H3>, это категория
                    print(f"Найдена категория: {'  ' * level}{h3_tag.text}")  # Отладочный вывод
                    output_file.write('\t' * level + h3_tag.text + '\n')  # Записываем категорию
                    # Ищем вложенный <DL> после <H3>
                    next_dl = h3_tag.find_next_sibling('dl')
                    if next_dl:  # Если нашли вложенный <DL>, обрабатываем его
                        print(f"Найден вложенный <DL> для категории: {h3_tag.text}")  # Отладочный вывод
                        parse_categories(next_dl, output_file, level + 1)
                    continue  # Переходим к следующему элементу
                elif a_tag:  # Если нашли <A>, это ссылка, а не категория
                    print(f"Найдена ссылка: {a_tag.text}, пропускаем")  # Отладочный вывод
                    continue  # Переходим к следующему элементу
                else:
                    print("Внутри <DT> не найден ни <H3>, ни <A>. Пропускаем.")  # Отладочный вывод
                    continue  # Переходим к следующему элементу
            else:
                print("Внутри <p> не найден <DT>")  # Отладочный вывод
                continue  # Переходим к следующему элементу
        elif element.name == 'dt':  # Если это <DT>
            h3_tag = element.find('h3')  # Ищем <H3> внутри <DT>
            a_tag = element.find('a')  # Ищем <A> внутри <DT>
            if h3_tag:  # Если нашли <H3>, это категория
                print(f"Найдена категория: {'  ' * level}{h3_tag.text}")  # Отладочный вывод
                output_file.write('\t' * level + h3_tag.text + '\n')  # Записываем категорию
                # Ищем вложенный <DL> после <H3>
                next_dl = h3_tag.find_next_sibling('dl')
                if next_dl:  # Если нашли вложенный <DL>, обрабатываем его
                    print(f"Найден вложенный <DL> для категории: {h3_tag.text}")  # Отладочный вывод
                    parse_categories(next_dl, output_file, level + 1)
                continue  # Переходим к следующему элементу
            elif a_tag:  # Если нашли <A>, это ссылка, а не категория
                print(f"Найдена ссылка: {a_tag.text}, пропускаем")  # Отладочный вывод
                continue  # Переходим к следующему элементу
            else:
                print("Внутри <DT> не найден ни <H3>, ни <A>. Пропускаем.")  # Отладочный вывод
                continue  # Переходим к следующему элементу
        elif element.name == 'dl':  # Если это вложенный <DL>
            print(f"Найден вложенный <DL>")  # Отладочный вывод
            parse_categories(element, output_file, level)  # Обрабатываем вложенный <DL>
        else:
            print(f"Пропущен элемент: {element.name}")  # Отладочный вывод

# Открываем и читаем файл HTML
try:
    with open('bookmarks_22.03.2025.html', 'r', encoding='utf-8') as file:
        html_content = file.read()
except Exception as e:
    print(f"Ошибка при чтении файла: {e}")
    exit()

# Создаем объект BeautifulSoup для парсинга HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Находим корневой <DL> (список закладок)
root_dl = soup.find('dl')  # BeautifulSoup игнорирует регистр
if not root_dl:
    print("Корневой <DL> не найден. Проверьте структуру HTML-файла.")
else:
    print("Корневой <DL> найден.")  # Отладочный вывод
    # Открываем файл для записи категорий
    try:
        with open('categories.txt', 'w', encoding='utf-8') as output_file:
            # Начинаем рекурсивный обход с корневого <DL>
            parse_categories(root_dl, output_file)
    except Exception as e:
        print(f"Ошибка при записи в файл: {e}")
        exit()
    print("Категории с указанием вложенности успешно записаны в файл categories.txt")
```
